```python
from math import gcd

def count_divisorial_points(l, r, n):
    if l + r > n:
        return 0
    if l + r == n:
        return 1
    return count_divisorial_points(l + r, r,n) + count_divisorial_points(l, l + r,n)
 
# from https://stackoverflow.com/a/18114286
def totient(n):
    amount = 0        
    for k in range(1, n + 1):
        if gcd(n, k) == 1:
            amount += 1
    return amount
```

```python
for i in range(1, 100):
    print (i, count_divisorial_points(1, 1, i), totient(i))
```
